```python
dados_vendas['Preco_m2'] = dados_vendas['Valor_anuncio'] / dados_vendas['Metragem']
dados_vendas
```

```python
metro_quadrado_bairro = dados_vendas.groupby('Bairro').agg(
    Valor_total = ('Valor_anuncio', 'sum'),
    M2_total = ('Metragem', 'sum')
).reset_index()

# Agora, calcule a média de m2 depois de ter feito o agrupamento
metro_quadrado_bairro['Media_m2'] = metro_quadrado_bairro['Valor_total'] / metro_quadrado_bairro['M2_total']

metro_quadrado_bairro
```

```python
top_bairros = dados_vendas['Bairro'].value_counts()[:10].index
```

```python
metro_quadrado_bairro.query('Bairro in @top_bairros')
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(data=metro_quadrado_bairro.query('Bairro in @top_bairros'), x='Bairro', y='Media_m2')
plt.title('Preço Médio por Metro Quadrado por Bairro')
plt.xlabel('Bairro')
plt.ylabel('Preço Médio por Metro Quadrado (R$)')
plt.xticks(rotation=45)
plt.show()
```

```python
quartis = dados_vendas['Valor_anuncio'].quantile()
quartis
```

```python
mediana = dados_vendas['Valor_anuncio'].median()
mediana
```

```python
Q1 = dados_vendas['Valor_anuncio'].quantile(0.25)
Q3 = dados_vendas['Valor_anuncio'].quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

outliers = dados_vendas[(dados_vendas['Valor_anuncio'] < limite_inferior) | (dados_vendas['Valor_anuncio'] > limite_superior)]
outliers
```

```python
plt.figure(figsize=(12, 12))
sns.boxplot(data=dados_vendas['Valor_anuncio'])
plt.title("Boxplot com Quartis, Mediana e Outliers")
plt.show()
```
