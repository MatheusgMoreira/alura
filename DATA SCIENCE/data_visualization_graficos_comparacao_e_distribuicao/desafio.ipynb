```python
import pandas as pd
url = 'https://raw.githubusercontent.com/afonsosr2/dataviz-graficos/master/dados/relatorio_vendas.csv'
```

```python
# Importando o relatório de vendas e atualizando as colunas de pedido e envio para o tipo data
vendas = pd.read_csv(url)
vendas['data_pedido'] = pd.to_datetime(vendas['data_pedido'], format='%Y-%m-%d')
vendas['data_envio'] = pd.to_datetime(vendas['data_envio'], format='%Y-%m-%d')

vendas
```

```python
# Verificando os tipos de dados e se existem dados nulos
vendas.info()
```

```python
df_lucro_ano = vendas.copy()
df_lucro_ano = df_lucro_ano[['data_pedido', 'lucro']]
df_lucro_ano
```

```python
df_lucro_ano['ano'] = df_lucro_ano.data_pedido.dt.year
df_lucro_ano.drop(labels='data_pedido', axis=1, inplace=True)
df_lucro_ano = df_lucro_ano.groupby('ano').agg('sum')
df_lucro_ano
```

```python
df_vendas_produtos = vendas.copy()
df_vendas_produtos = df_vendas_produtos[['data_pedido', 'tipo_produto', 'vendas']]
df_vendas_produtos
```

```python
df_vendas_produtos['ano'] = df_vendas_produtos.data_pedido.dt.year
df_vendas_produtos.drop(labels='data_pedido', axis=1, inplace=True)
```

```python
df_vendas_produtos = df_vendas_produtos.groupby('tipo_produto')[['vendas']].sum()
df_vendas_produtos
```
